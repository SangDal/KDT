#1.Clusters

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs # 데이터셋을 만들어주는 애
```

```python
# X, y값을 랜덤으로 만들어줌.
X, y = make_blobs(n_samples=100, centers=3, random_state= 10)
```

```python
# X값(문제의값) 확인 (100개의 row가 보임)
X = pd.DataFrame(X)
X
```

```python
# y(답의값)
y
```

```python
sns.scatterplot(x=X[0], y=X[1], hue=y)
```

```python
from sklearn.cluster import KMeans
```

```python
# 비지도로  한번 해봅시다.
# KMeans에게 답을 모르겠으니 3이라는 숫자를 넣고 한번 해봐
km = KMeans(n_clusters=3)
```

```python
km.fit(X)
```

```python
# 예측값
pred = km.predict(X)
```

```python
sns.scatterplot(x=X[0], y=X[1], hue=pred)
```

```python
# 값을 바꿔서 한번 해봅시다.
km = KMeans(n_clusters=4)
km.fit(X)
pred = km.predict(X)
```

```python
sns.scatterplot(x=X[0], y=X[1], hue=pred)
```

```python
# km 평가값 : 하나의 클러스터 안에 중심점으로부터 각각의 데이터 거리를 합한값.
km.inertia_
```

```python
inertia_list = []

# 2~10까지 일단 실행 하본다.
for i in range(2,11):
  km = KMeans(n_clusters=i)
  km.fit(X)
  inertia_list.append(km.inertia_)
```

```python
# 급격하게 꺽기는 부분이 최적의 K값이라고 보면 된다 !!
sns.lineplot(x= range(2,11), y= inertia_list)
```

#3.스케일링(Scaling)
* 데이터를 특정한 스케일로 통일하는 것을 말한다.
* 데이터를 모델링 하기 전 거치는 것이 좋다
* 다차원의 값들을 비교 분석하기 쉽게 만들어주며, 자료의 오버플로우나 언더 플로우를 방지하여 최적화 과정에서의 안정성 및 수렴 속도를 향상


###3-1. 스케일링의 종류
* StandardScaler: 평균과 표준편차를 사용
* MinMaxScaler: 최대/최소값이 각각 1과 0이 되도록 스케일링
* RobustScaler: 중앙값과 IQR사용, 아웃라이어의 영향을 최소화

#4.KMeans
* KMeans 클러스터링 알고리즘은 n개의 중심점을 찍은 후에, 이 중심점에서 각 점간의 거리의 합이 가장 최소화가 되는 중심점 n의 위치를 찾고, 이 중심점에서 가까운 점들을 중심점을 기준으로 묶는 알고리즘 !
* 중심점은 각 군집의 데이터의 평균값을 위치로 가지게 됨
* 클러스터의 수를 사람이 정해줘야 함
* 중심점을 측정할떄 처음에는 랜덤으로 중심점의 위치를 찾기 때문에 잘못하면 중심점의 점간의 거리가 Global optimum인 최소값을 찾는 게 아니라 중심점이 Local optimum에 수렴하여 잘못된 분류를 할 수 있음

```python
inertia_list = []

# 2~10까지 일단 실행해본다.
for i in range(2,11):
  km = KMeans(n_clusters=i, random_state=10)
  km.fit(mm_df)
  inertia_list.append(km.inertia_)
```

```python
inertia_list
```

```python
# 드라마틱한 엘보우가 나오지 않는다.
sns.lineplot(x=range(2,11), y=inertia_list)
```

#5.시루엣스코어
* 각 군집 간의 거리가 얼마나 효율적으로 돼있는지리를 나타낸다.
* 다른 군집과의 거리는 떨어져 있고 동일 군집끼리의 데이터는 서로 가깝게 잘 뭉쳐 있다는 의미
* 실루엣 분석은 실루엣 계수를 기반으로 하는데, 실루엣 계수는 개별 데이터가 가지는 군집화 지표

```python
from sklearn.metrics import silhouette_score
```

```python
score = []

# 2~10까지 일단 실행해본다.
for i in range(2,11):
  km = KMeans(n_clusters=i, random_state=10)
  km.fit(mm_df)
  pred = km.predict(mm_df)
  score.append(silhouette_score(mm_df, pred))
```

```python
# 스코어 값이 높으면 높을수록 좋다.
score
```

```python
# 가장 이상적인 값은 '8'
sns.lineplot(x=range(2,11), y=score)
```

```python
# 위에서 찾은 최고의 값을 주고 다시 해보기
km = KMeans(n_clusters=8, random_state=10)
```

```python
# 학습
km.fit(mm_df)
```

```python
pred = km.predict(mm_df)
```
