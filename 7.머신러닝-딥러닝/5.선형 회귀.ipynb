### 데이터 칼럼 알아보기

* Posted On: 매물 등록 날짜
* BHK: 베드, 홀, 키친의 개수
* Rent: 렌트비
* Size: 집 크기
* Floor: 총 층수 중 몇층
* Area Type: 공용공간을 포함하는지, 집의 면적만 포함하는지
* Area Locality: 지역
* City: 도시
* Furnishing Status: 풀옵션 여부
* Tenant Preferred: 선호하는 가족형태
* Bathroom: 화장실 개수
* Point of Contact: 연락할 곳
