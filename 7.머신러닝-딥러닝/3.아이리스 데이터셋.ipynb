```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
iris
```

```python
print(iris['DESCR'])
```

#데이터의 설명

- sepal length in cm :  꽃받침의 길이
- sepal width in cm : 꽃받침의 너비
- petal length in cm : 꽃잎의 길이
- petal width in cm : 꽃잎의 너비

- class:
        - Iris-Setosa
        - Iris-Versicolour
        - Iris-Virginica

```python
data = iris['data']
data
```

```python
target = iris['target']
target
```

```python
feature_names = iris['feature_names']
feature_names
```

```python
import pandas as pd
```

```python
df_iris = pd.DataFrame(data, columns=feature_names)
df_iris.head()
```

```python
df_iris['target'] = target
df_iris
```

```python
from sklearn.model_selection import train_test_split
```

```python
# train_test_split(독립변수, 종속변수, 테스트사이즈=0.25, 시드값)
# 테스트사이즈는 디폴트 0.25

#순서를 지켜줘야한다 !
X_train, X_test, y_train, y_test = train_test_split(df_iris.drop('target',1),
                                                    df_iris['target'],
                                                    test_size=0.2,
                                                    random_state=10)
```

```python
X_train.shape, X_test.shape 
```

```python
y_train.shape, y_test.shape
```

```python
X_train
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score # accuracy_score : 얼마나 정확히 맞췄는데 확인하는애
```

```python
svc = SVC()
```

```python
svc.fit(X_train, y_train)
```

```python
y_pred = svc.predict(X_test)
```

```python
# 똑같은 데이터로 하여도 썪이는거에 따라서 정답률은 모든 사람이 다를수 있다(강사님이랑 값이 같을수도 다를수도 있음)
print('정답률 : ',accuracy_score(y_test, y_pred))
```
