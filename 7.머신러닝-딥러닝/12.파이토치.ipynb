#1.파이토치(pytorch)
* 텐서플로우와 함께 머신러닝, 딥러닝에서 가장 널리 사용되는 프레임워크

```python
import torch
#버전확인
print(torch.__version__)
```

###1-1.스칼라(Scalar)
* 하나의 상수를 의미한다.


```python
# torch에서는 tensor로 상수를 담는다.
# tensor안에 있는 상수 '1'이 스칼라이다.
var1 = torch.tensor([1])
var2 = torch.tensor([6.5])
```

```python
type(var1)
print(var1)
print(var2)
```

```python
# 두 스칼라의 사칙 연산
print(var1 + var2)
print(var1 - var2)
print(var1 * var2)
print(var1 / var2)
```

###1-2.벡터(Vector)
* 상수가 두개 이상 나열되었을 경우 벡터라고 한다.

```python
vector1 = torch.tensor([1,2,3])
print(vector1)
vector2 = torch.tensor([10,20,30])
print(vector2)
```

```python
# 두 벡터의 사칙 연산
print(vector1 + vector2)
print(vector1 - vector2)
print(vector1 * vector2)
print(vector1 / vector2)
```

###1-3.행렬(Matrix)
* 2개 이상의 벡터 값을 가지고 만들어진 값으로 행과 열의 개념을 가진 숫자의 나열

```python
matrix1 = torch.tensor([[1,2],[3,4]])
print(matrix1)
```

```python
matrix2 = torch.tensor([[7,8],[9,10]])
print(matrix2)
```

```python
# 두 행렬의 사칙 연산
print(matrix1 + matrix2)
print(matrix1 - matrix2)
print(matrix1 * matrix2)
print(matrix1 / matrix2)
```

###1-4.텐서(Tensor)
* 다수의 행렬이 모이면 텐서라고 부른다...
* 배열이나 행렬과 매우 유사한 특수 자료구조이다.
* 파이토치는 텐서를 사용하여 모델의 입력과 출력 , 모델의 매개변수들을 처리하는데 사용됨

```python
from IPython.display import Image
Image(url='https://miro.medium.com/max/875/1*jRyyMAhS_NZxqyv3EoLCvg.png')
```

```python
tensor1 = torch.tensor([[[1,2],[3,4]],[[5,6],[7,8]]])
print(tensor1)
```

```python
tensor2 = torch.tensor([[[9,10],[11,12]],[[13,14],[15,16]]])
print(tensor2)
```

```python
# 두 텐서의 사칙 연산
print(tensor1 + tensor2)
print(tensor1 - tensor2)
print(tensor1 * tensor2)
print(tensor1 / tensor2)
```

```python
print(torch.add(tensor1, tensor2))      # +
print(torch.subtract(tensor1, tensor2)) # -
print(torch.multiply(tensor1, tensor2)) # *
print(torch.divide(tensor1, tensor2))   # /
print(torch.matmul(tensor1, tensor2))   # 행렬곱
```

```python
# 앞과 뒤를 더한후 재저장해준다.
# 모든 사칙연산자의 _를 붙여 사용할 수 있음

# tensor1 + tensor2 한후 tensor1에 다시 저장해라
print(tensor1.add_(tensor2))
print(tensor1)
```

#2.텐서의 변환

```python
# 일반적인 리스트
data = [[1,2], [3,4]]
print(data)
print(type(data))
```

```python
# 일반적인 리스트를 텐서로 바꾸기
x_data = torch.tensor(data)
type(x_data)
```

```python
import numpy as np
```

```python
# 엔디어레이도 텐서로 변경 가능하다.
np_array = np.array(data)
np_array
```

```python
#np_array를 tensor로 바꾸는방법
x_np_1 = torch.tensor(np_array)
x_np_1
```

```python
# 텐서 안에 있는 요소의 값도 변경 가능하다.

x_np_1[0,0] =100
print(x_np_1)

print(np_array)
```

```python
x_np_2 = torch.as_tensor(np_array) #np_array와 동일한 메모리 주소를 가리키는 뷰를 만드는 함수
print(x_np_2)

x_np_2[0,0] = 200 # 기존 메모리 주소의 ndarragy 값을 변경하게 됨
print(x_np_2)
print(np_array)
```

```python
# 엔디어레이만 변환 해줍니다.
x_np_3 = torch.from_numpy(np_array)#np_array와 동일한 메모리 주소를 가리키는 뷰를 만드는 함수
print(x_np_3)

x_np_3[0,0] = 400 # 기존 메모리 주소의 ndarray값을 변경하게 됨
print(x_np_3)
print(np_array)
```

```python
# 엔디어레이도 되돌리기
np_again = x_np_1.numpy()
print(np_again,type(np_again))
```

#3.파이토치 주요 함수


```python
# 2행 3열 을 모두 1로 채워준다.
a = torch.ones(2,3)
print(a)
```

```python
#2행 3열 짜리 행렬을 만들고 0으로 모두채움
b= torch.zeros(2,3)
print(b)
```

```python
# 2행 3열 짜리 행렬을 만들고 원하는 숫자로 채워진다.
c = torch.full((2,3), 10)
print(c)
```

```python
# 2행 3열 짜리 행렬을 만들고 실수 쓰레기값을 넣어준다.
d= torch.empty(2,3)
print(d)
```

```python
# 무조건 정사각형 행렬을 생성해주고 대각선을 '1'로 채워준다.
e= torch.eye(5)
print(e)
```

```python
# 일반적이 arange랑 같다.
# 0~9까지 순차적으로 뽑기
f=torch.arange(10)
print(f)
```

```python
# 0~1사이의 값을 랜덤으로 뽑아준다.
g= torch.rand(2,3)
print(g)
```

```python
# -1~0사이의 값을 랜덤으로 뽑아준다.
h= torch.randn(2,3)
print(h)
```

```python
# reshape(2,2,4): 2개의 행렬이고 2행4열이다.
i = torch.arange(16).reshape(2,2,4)
print(i,i.shape)

# transpose(): 차원을 인덱스로 변경
j =i.transpose(1,2) # 2,2,4 -> 2,4,2 로 변환 2개의 행렬이고 4행 2열
print(j, j.shape)
```

```python
# permute(): 차원을 인덱스로 변환
k = i.permute((2,0,1)) # 원래 있던 2,2,4 를 4,2,2로 변환 시켜주는것
print(k, k.shape)
```

#4.GPU 사용하기
* 코랩에서 device 변경하는 방법
  * 상단메뉴 -> 런타임 -> 런타임 유형변경 -> 하드웨어 가속기를 GPU로 변경 -> 저장 -> 다시시작 및 모두 실행

```python
tensor = torch.rand(3,4)
print(f'shape:{tensor.shape}')
print(f'type:{tensor.dtype}')
print(f'device :{tensor.device}')
```

```python
# cuda : GPU를 의미한다.
# is_available: 사용가능하냐? 판단
tensor = tensor.reshape(4,3)
tensor = tensor.int()
if torch.cuda.is_available():
  print('GPU를 사용할 수 있음')
  tensor = tensor.to('cuda') # GPU로 보내는 역할을함

print(f'shape:{tensor.shape}')
print(f'type:{tensor.dtype}')
print(f'device :{tensor.device}')
```
