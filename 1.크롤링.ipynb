* 크롤링(Crawling): 인터넷의 데이터를 활용하기 위해 정보들을 분석하고 활용할수 있도록 수집하는 행위를 말한다.

#1.Basic English Speaking

https://basicenglishspeaking.com/daily-english-conversation-topics/

```python
import requests
from bs4 import BeautifulSoup #파싱 할때 쓰이는 모듈
```

```python
#변수에 사이트 저장
site = 'https://basicenglishspeaking.com/daily-english-conversation-topics/'

request = requests.get(site) # requests 모듈을 이용해 get방식으로 위에 저장한 사이트로 보낸다.
print(request) # request변수안에는 객체로 request status 코드가 존재한다. 
# print(request.text) # request.text란 사이트의 모든 코드를 가져온다. 
```

```python
soup = BeautifulSoup(request.text)
print(soup)
```

```python
# find() : 단일 테그만 가져 올 수 있음 
# find('테그이름',{키값:벨류})
divs = soup.find('div',{'class':'thrv-columns'})
print(divs)
```

```python
# 위에서 불러온 div박스 안에있는 모든 a테그 값을 가져온다.
links = divs.findAll('a')
print(links)
```

```python
# links에서 불러온 a테그의 내용에서 text만 다 찍어주세요.
for link in links:
  print(link.text)
```

```python
# subject라는 배열을 생성해서 위에서 뽑아온 text내용을 subject에 저장한다(배열로)
subject = []
for link in links:
  subject.append(link.text)
```

```python
len(subject)
```

```python
print('총', len(subject), '개의 주제를 찾았습니다. ')
for i in range(len(subject)):
  print('{0:2d}, {1:s}'.format(i+1, subject[i]))
```

#2.다음 뉴스 기사 
* 제목을 크롤링하기 

```python
#뉴스 기사 URL 따오기 
# 뉴스: https://v.daum.net/v/20230601090108867
# 영화: https://v.daum.net/v/20230531203030159

def daum_news_title(new_id):
  url = 'https://v.daum.net/v/{}'.format(new_id)
  request = requests.get(url)
  soup = BeautifulSoup(request.text)
  title = soup.find('h3', {'class': 'tit_view'})
  if title:
    return title.text.strip()
  return '제목없음'

```

#3.벅스뮤직
* 벅스 차트에서 노래명과 아티스트 크롤링
* https://music.bugs.co.kr/chart

```python
request =  requests.get('https://music.bugs.co.kr/chart')
soup = BeautifulSoup(request.text)
```

```python
# 벅스뮤직에 헤더 담기
header = {'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64)'}
request = requests.get('https://music.bugs.co.kr/chart', headers = header)
print(request)
```

```python
titles = soup.findAll('p', {'class' : 'title'})
print(titles) #잘 크롤링 되었는지 확인
artists = soup.findAll('p', {'class' : 'artist'})
# print(artists) #잘 크롤링 되었는지 확인
```

```python
for i in range(len(titles)):
  title = titles[i].text.strip()
  artist = artists[i].text.strip().split('\n')[0]
  print('{0:3d}위 {1} - {2}'.format(i+1, artist, title))
```

```python
# zip함수로 변경해보기
# 인터러블한 객체를 여러개 넣을수 있는데 안에 데이터의 갯수가 다를수 있다. 그땐 제일 짧은애 기준으로 한다.
# enumerate(): 인덱스를 뽑아 낼수 있는 함수
# i = enumerate에서 나오는 인덱스 
# t = titles
# a = artists

for i, (t, a) in enumerate(zip(titles, artists)):
  title = t.text.strip()
  artist = a.text.strip().split('\n')[0]
  print('{0:3d}위 {1} - {2}'.format(i+1, artist, title))
```

#5.네이버주식
* https://finance.naver.com/item/main.naver?code=005930


```python
site = 'https://finance.naver.com/item/main.naver?code=005930'
request = requests.get('https://finance.naver.com/item/main.naver?code=005930')
print(request)
```

```python
soup = BeautifulSoup(request.text)
```

```python
div_today = soup.find('div', {'class' : 'today'})
print(div_today)
```

```python
em = div_today.find('em')
print(em)
```

```python
price = em.find('span', {'class' : 'blind'}).text
print(price)
```

```python
# 종목이름 크롤링
wrap_company = soup.find('div',{'class':'wrap_company'})
# print(wrap_company)
name = wrap_company.a.text
print(name)
```

```python
div_description = wrap_company.find('div', {'class':'description'})
# print(div_description)
code = div_description.span.text
print(code)
```

```python
# 거래량 뽑기 

ta_info = soup.find('table', {'class':'no_info'})
# print(ta_info)
tds = ta_info.find_all('td')
# print(tds)
volume = tds[2].find('span', {'class':'blind'}).text
print(volume)
```

```python
dic = {'price':price, 'name':name, 'code':code, 'volume':volume}
dic
```

```python
# 네이버 주식창을 이용하여 코드를 입력받으면 딕셔너리 값으로 출력 되는 함수 만들기 !
def naver_finanace(code):
  site = f'https://finance.naver.com/item/main.naver?code={code}'
  request = requests.get(site)
  soup = BeautifulSoup(request.text)
  div_today = soup.find('div', {'class' : 'today'})
  em = div_today.find('em')

  #가격
  price = em.find('span', {'class' : 'blind'}).text

  #회사명
  wrap_company = soup.find('div',{'class':'wrap_company'})
  name = wrap_company.a.text

  #코드
  div_description = wrap_company.find('div', {'class':'description'})
  code = div_description.span.text

  # 거래량 뽑기 
  ta_info = soup.find('table', {'class':'no_info'})
  tds = ta_info.find_all('td')
  volume = tds[2].find('span', {'class':'blind'}).text

  # 딕셔너리로 보여주기 
  dic = {'price':price, 'name':name, 'code':code, 'volume':volume}
  return dic
```

```python
naver_finanace('016740')
```

```python
codes = ['005930','000660', '035420', '005950']
```

```python
data = []

for code in codes:
  dic = naver_finanace(code)
  data.append(dic)
print(data)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(data)
df
```

```python
df.to_excel('naver_finace.xlsx')
```

#과제.지니차트200


```python
# 지니 차트 200 위까지 크롤링하기 
# 4번 하는게 아니라 반복문을 이용하기
# 수집한 데이터를 데이터프레임으로 저장하기
# 엑셀로 출력해보기 
# github에 올려두기 
```

```python
import requests
from bs4 import BeautifulSoup #파싱 할때 쓰이는 모듈
```

```python
# https://www.genie.co.kr/chart/top200?ditc=D&ymd=20230601&hh=21&rtm=Y&pg=1
# 1~50 첫페이지만 먼저 TEST
genie_site = "https://www.genie.co.kr/chart/top200?ditc=D&ymd=20230601&hh=21&rtm=Y&pg=1"


header = {'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0.0.0 Safari/537.36'}
request = requests.get(genie_site, headers = header)
soup = BeautifulSoup(request.text)
print(request)
```

```python
tables = soup.find('table', {'class' : 'list-wrap'})
# print(tables)
genie_titles = tables.findAll('a', {'class':'title'})
#print(genie_titles)
genie_artists = tables.findAll('a', {'class' : 'artist'})
#print(genie_artists)
```

```python
for i, (gt, ga) in enumerate(zip(genie_titles, genie_artists)):
  genie_title = gt.text.strip()
  genie_artist = ga.text.strip()
  print('{0:3d}위 {1} - {2}'.format(i+1, genie_title, genie_artist))
```

```python
# 1~200위까지 뽑기 !
page = ['1','2','3','4']
k = 1
# data 엑셀 만들기 위한 데이터
data = []

for i in page:
  genie_site = f"https://www.genie.co.kr/chart/top200?ditc=D&ymd=20230601&hh=21&rtm=Y&pg={i}"

  header = {'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0.0.0 Safari/537.36'}
  request = requests.get(genie_site, headers = header)
  soup = BeautifulSoup(request.text)
  print(f"\n{i}번째 페이지")

  tables = soup.find('table', {'class' : 'list-wrap'})
  genie_titles = tables.findAll('a', {'class':'title'})
  genie_artists = tables.findAll('a', {'class' : 'artist'})
  

  for j, (gt, ga) in enumerate(zip(genie_titles, genie_artists), start=k):
    genie_title = gt.text.strip()
    genie_artist = ga.text.strip()
    print('{0:3d}위 {1} - {2}'.format(j, genie_title, genie_artist))
    dic = {'ranking': k, 'title': genie_title, 'artist': genie_artist}
    data.append(dic)
    k= k+1
```

```python
# 엑셀 파일로 저장하기
import pandas as pd
```

```python
print(data)
```

```python
df = pd.DataFrame(data)
df
```

```python
df.to_excel('genie1~200.xlsx')
```
