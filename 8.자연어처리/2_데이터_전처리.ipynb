#2.토큰화(Tokenization)
* 전처리 과정에는 토큰화, 정제, 정규화 작엄등이 있다.
* 이때, 토큰화는 원시 데이터를 가져와 유용한 데이터 문자열로 변환하는 간다한 프로세스 입니다.
* 원래 토큰은 사이버 보안, NFT 생성에 사용되는 것으로 유명, 자연어 프로세스의 중요한 부분을 차지한다.
* 토큰의 단위는 자연어 내에서 의미를 가지는 최소 단위로 정의 할 수 있다.(모델마다 다를수 있다.)
* 단락과 문장을 보다 쉽게 의미를 할당할 수 있는 더 작은 단위로 분할하는데 사용된다.

> 결론적으로 토큰화 작업은 주어진 코퍼스내 자연어 문장들을 토큰이라 불리는 최소 단위로 나누는 작업이다.

###2-1.토큰화 과정의 필요성
* 언어 모델의 자연어 이해 능력을 향상시키기 위함이다.
---
<모델마다 끊는 방법이 다르다>

    "역삼동에 어떤 맛집이 있나요?"
    "역삼동에" "어떤" "맛집이" "있나요?"
    "역삼동" "에" "어떤" "맛집" "이" "있나" "요?"
---
* 다양한 자연어를 효율적으로 표현 가능하다.
    - 한글은 자음 19개, 모음 21개
    - 중복을 제거한 대규모 자연어 코퍼스 내 토큰의 집합(단어 사전) 사용
    - 적절한 정보량을 내포하면서 전체 단어 사전의 개수가 많아지지 않도록 토큰의 단위를 잘 정의해야 한다.
    - 상황에 따라 다르지만, 일번적인 자연어 처리 작업에서 단어 사전의 규모는 약 10,000 ~ 50,000개 정도로 구성(현대 한글의 글자수는 11,172개)

###2-2.토큰화 방법
* 문장 토큰화: 토큰의 기준을 문장으로 하는 토큰화 방법
    - 문장의 끝에 오는 문장 부호를 기준으로 코퍼스를 잘라낸다.(. 또는 ! 또는 ?)


    - 예외)
    "제 이메일 주소는 skckdeo@naver.com입니다." 일경우
    "제 이메일 주소는 skckdeo@naver" ".com입니다." 이렇게 짤라버림.

* 단어 토큰화: 토큰의 기준을 단어로 하는 토큰화 방법
    - 보편적으로 구분기호를 가지고 텍스트를 나누게 되며, 기본적으로 공백을 구분자로 사용한다.
    - 한국어의 경우 교착어이기 떄문에 고백으로 단어 토큰화를 하면 성능이 좋지 않다.
    - 새로운 단어가 추가될수록 단어 사전의 크기가 계속 증가
    - OOV(Out Of Vocabulary)문제가 발생할 수 있다.
* 문자 토큰화: 토큰의 기준을 문자로 하는 토큰화 방법
    - 단어 토큰화의 한계점들을 해결하기 위한 방법이다.
    - 영어는 26개의 알파벳에 따라 분리, 그러나 한국어는 자음 19개와 모음 21개의 글자에 따라 분리한다.
    - 문장 하나를 생성하는데 너무 많은 추론이 필요하다.
    - 단어 사전은 작지만 모델의 예측 시간에 문제가 생길 수 있다.

* 서브워드 토큰화: 토큰의 기준을 서브워드로 하는 토큰화방법(지금 현재 구글에서 만든 알고리즘을 가장 많이 쓴다)
    - 단어 토큰화와 문자 토큰화의 한계점을 해결하기 위한 방법
    - 문자 토큰화의 확장된 버전으로 토큰의 단위를 n개의 문자로 정의하고, 해당 기준에 따라 텍스트를 분절하는 방법
    - 형태소 분절 기반의 서브워드 토큰화로 확장 될 수 있어 한국어에서도 좋은 성능을 가짐
    - 서브워드를 만드는 알고리즘 중에서 가장 유명한 것이 BPE
    

#3.서브워드 토큰화

###3-1.서브워드(subword)
* 단어보다 더 작은 의미의 단위
* 예) Birthday = birth + day, 아침밥 = 아침 + 밥
* 단어를 여러 서브워드로 분리해서 단어 사전을 구축하겠다는 토큰화 방법
* 신조어에서 주로 발생하는 OOV 문제를 완화

#과제
"BPE알고리즘", "WordPiece Tokenizer" 알고리즘에 대해 조사하고 해당 알고리즘을 잘 설명할 수 있는 예제와 실급과제를 완성해보자.

###3-2.BPE(Byte Pair Encoding)
* 등장 빈도!
* 코퍼스 내 단어의 등장 빈도에 따라 서브워드를 구축하는데 사용한다.
* 2016년 [Neural Machine Translation of Rare Words With Subword Units](https://arxiv.org/abs/1508.07909) 논문에서 처음 제안
* BPE를 한문장으로 압축하면 : 글자단위에서 점진적으로 서브워드 집합을 만들어내는 Botton-up 방식의 접근 방식으로 자연어 코퍼스에 있는 모든 단어들을 글자 단위로 분리한 뒤에, 등장빈도에 따라 글자들을 서브워드로 통합하는 방식이다.



###3-3.WordPiece Tokenizer
* 구글이 2016년도 [Google's Neural Mashine Translation System Bridging the Gap betwwn Human and Machine Translation](https://arxiv.org/abs/1609.08144) 논문에 처음 공개한 BPE의 변형 알고리즘
* 병합할 두 문자가 있을 때, 각각의 문자가 따로 있을 때를 더 중요시 여기는지, 병합되었을 때를 더 중요시 여기는지에 차이점을 둠
* GPT 모델과 같은 생성 모델의 경우에는 BPE 알고리즘을 사용 한다
* BERT, ELECTRA와 같은 자연어 이해 모델에서는 WordPiece Tokenizer를 주로 사용한다.

###5.정규화
* 핵심키워드 범위! 범위를 선택한다!

###5-1.정규화 작업이 필요한 이유
* 이상적으로 단어 사전내의 단어 토큰들이 모두 중요하게 고려되길 원함도 있다.

* 의미가 같지만 표기가 다른 단어들을 통합할 수 있다면 통합된 단어의 사용 빈도가 높아질것이고, 빈도가 낮은 단어들의 중요도가 높아질 수 있음

###5-2.정규화 작업의 종류

    - 포제어 추출: 단어들이 다른형태를 가지더라도, 그 뿌리 단어를 찾아가서 단어의 개수를 줄일 수 있는지 판단하는 방법
* 대소문자 통합
    - 모든 단어를 소문자로 통합해서 사용하는 편이다. 예외가 존재 하기도한다!
    - 대문자와 소문자가 구분되어야 하는 경우도 존재한다, 따라서 무작정 소문자로 통합해서는 안되는경우가 있음 예) US(미국), us(우리)

###7-1.불용어 제거

```python
# 불용어 사전 정의
stopword = ['이하', '바로','☞','※','…']

# 불용어 제거 함수
def delete_stopwords(context):
    preprocessed_text = []
    for text in context:
        text = [w for w in text.split(' ') if w not in stopword]
        preprocessed_text.append(' '.join(text))
    return preprocessed_text
```

```python
processed_context = delete_stopwords(context)
```

```python
for i, text in enumerate(processed_context):
    print(i, text)
```

###7-2.HTML 태그 제거

```python
#정규식을 쓸수 있는 모듈

import re
```

```python
def delete_html_tag(context):

    preprocessed_text = []

    for text in context:
        # 정규식 패턴을 사용하여 HTML 태그 제거
        # 예시) re.sub(r'정규식', '', text).strip()
        text = re.sub(r'<[^>]+>\s+(?=<)|<[^>]+>', '', text).strip()
        if text:
            preprocessed_text.append(text)
    return preprocessed_text
```

```python
processed_context = delete_html_tag(context)
```

```python
for i, text in enumerate(processed_context):
    print(i, text)
```

###7-3.문장 분리

```python
def delete_hashtag(context):

    preprocessed_text = []

    for text in context:
        # '#\S+':  "#뒤에 모두 지워잘라" 샵역슬러스대문자에스
        text = re.sub(r'#\S+', '', text).strip()
        if text:
            preprocessed_text.append(text)

    return preprocessed_text
```

###7-6.저작권 관련 테스트 제거


```python
def delete_copyright(context):
    re_patterns = [
        r'\<저작권자(\(c\)|©|(C)|(\(C\))).+?\>',
        r'(Copyrights)|(\(c\))|(\(C\))|©|(C)|'
    ]
    preprocessed_text = []
    for text in context:
        for re_pattern in re_patterns:
            text = re.sub(re_pattern, "", text).strip()
        if text:
            preprocessed_text.append(text)
    return preprocessed_text
```

```python
preprocessed_text = delete_copyright(preprocessed_text)
```

```python
for i, text in enumerate(preprocessed_text):
    print(i, text)
```

```python
# Mecab은 속도가 굉장히 우수하다.
from konlpy.tag import Mecab
```

```python
mecab = Mecab()
morphs = mecab.pos('아버지가방에들어가신다.', join=False)
print(morphs)
```

```python
# 명사(NN), 동사(V), 형용사(J)의 포함 여부에 따라 문장 필터링
def morph_filter(context):
    NN_TAGS = ['NNG', 'NNP', 'NNB', 'NP']
    V_TAGS = ['VV', 'VA', 'VX', 'VCP', 'VCN', 'XSN', 'XSA', 'XSV']
    J_TAGS = ['JKS', 'J', 'JO', 'JK', 'JKC', 'JKG', 'JKB', 'JKV', 'JKQ',
                'JX', 'JC', 'JKI', 'JKO', 'JKM', 'ETM']
    preprocessed_text = []

    for text in context:
        morphs = mecab.pos(text, join=False)
        nn_flag = False
        v_flag = False
        j_flag = False

        for morph in morphs:
            pos_tags = morph[1].split("+")

            for pos_tag in pos_tags:
                if not nn_flag and pos_tag in NN_TAGS:
                    nn_flag = True
                if not v_flag and pos_tag in V_TAGS:
                    v_flag = True
                if not j_flag and pos_tag in J_TAGS:
                    j_flag = True

            if nn_flag and v_flag and j_flag:
                preprocessed_text.append(text)
                break
    return preprocessed_text
```

```python
post_precessed_context = morph_filter(normalizer_context)
```

```python
for i, text in enumerate(post_precessed_context):
    print(i, text)
```

###7-10.문장 길이 기반 필터링

```python
# 글의 최소 길이와 최대길이를 입력받아 문장 길이를
def min_max_filter(min_len, max_len, context):
    precessed_context = []
    for text in context:
        if min_len < len(text) and len(text) < max_len:
            precessed_context.append(text)
    return precessed_context
```

```python
post_precessed_context = min_max_filter(20, 60, post_precessed_context)
```

```python
for i, text in enumerate(post_precessed_context):
    print(i, text)
```
