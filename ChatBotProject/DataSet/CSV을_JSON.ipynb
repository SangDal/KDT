```python
import pandas as pd
import json
```

```python
df = pd.read_csv('/content/drive/MyDrive/프로젝트(web)/data/train_data.csv')
```

```python
df = df.drop('NO.', axis=1)
```

```python
df.dropna(inplace=True)
```

```python
df
```

```python
# 전체 데이터의 개수를 구합니다.
total_samples = len(df)

# 훈련 데이터 개수 계산 (전체 데이터의 80%)
train_samples = int(0.8 * total_samples)

# 데이터 랜덤
shuffled_df = df.sample(frac=1, random_state=42)

# 훈련 세트와 테스트 세트로 데이터를 나눕니다.
train_data = shuffled_df[:train_samples]
test_data = shuffled_df[train_samples:]

# 나눈 데이터를 확인합니다.
print("Train 데이터 개수:", len(train_data))
print("Test 데이터 개수:", len(test_data))
```

```python
train_data
```

```python
test_data
```

```python
json_data = {
    "version": "KorQuAD_v1.0_dev",
    "data": []
}

for index, (context, question) in enumerate(zip(test_data['자소서'], test_data['질문'])):
    json_data["data"].append({
        "paragraphs": [
            {
                "qas": [
                    {
                        "answers": [
                            {
                                "text": "?",
                                "answer_start": 0
                            }
                        ],
                        "id": index,
                        "question": question
                    }
                ],
                "context": context
            }
        ],
        "title": ""
    })
```

```python
json_data = json.dumps(json_data)
```

```python
with open('test_set.json', 'w', encoding='euc-kr') as f:
    f.write(json_data)
```
