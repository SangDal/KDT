```python
def convert_to_unicode_escape(original_text):
    escaped_text = original_text.encode('unicode_escape').decode()
    return escaped_text

# 사용 예시
if __name__ == "__main__":
    original_text = input("변환할 텍스트를 입력하세요: ")
    escaped_text = convert_to_unicode_escape(original_text)
    print("변환된 결과: ", escaped_text)
```

```python
import re

def convert_from_unicode_escape(escaped_text):
    def decode_match(match):
        return bytes(match.group(0), 'utf-8').decode('unicode_escape')

    pattern = re.compile(r'\\u[0-9a-fA-F]{4}')
    original_text = pattern.sub(decode_match, escaped_text)
    return original_text

# 사용 예시
if __name__ == "__main__":
    escaped_text = input("변환할 unicode_escape 텍스트를 입력하세요: ")
    original_text = convert_from_unicode_escape(escaped_text)
    print("변환된 결과: ", original_text)
```
